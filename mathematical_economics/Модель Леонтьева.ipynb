### Задача 1

#### Найти коэффициенты прямых материальных затрат.

```python
matrix = np.array([[40, 180, 25],
                  [160, 90, 250],
                  [80, 450, 50]])
Y = np. array([21, 100, 75])

X = np.sum(matrix, axis=1) + Y
print(f'Вектор валовых выпусков по отраслям равен {X}')
```

```python
 A = matrix / X
print(f'Матрица А коэффициентов прямых затрат выглядит так: \n\n {A}')
```

#### Вычислить объемы конечного продукта при увеличении валового выпуска каждой отрасли соответственно на 10%, 50% и 20%

```python
I = np.eye(3)
rise = [1.1, 1.5, 1.2]
new_Y = (I - matrix / (X * rise))@(X * rise)

print(f'Вектор конечных продуктов при увеличении валового выпуска {new_Y}')
```

### Задача 2

```python
A = np.array([[0.2, 0.1, 0.3],
              [0.3, 0.2, 0.1],
              [0.1, 0.3, 0.2]])

I = np.eye(3)

Y = np.array([310, 70, 20])

print(f'Валовые выпуски отраслей равны соответственно {np.linalg.solve((I - A), Y)}')
```
